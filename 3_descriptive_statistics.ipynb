**Measure of Central tendency**

```
#importing basic libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#dataframe - dataet into tabular column and row form
dataframe = pd.DataFrame(dict(id=range(6), age=np.random.randint(18,31, size=6)))
dataframe
```

```
#central tendency
print('mean for df',dataframe["age"].mean())
print('median for df',dataframe["age"].median())
print('mode for df',dataframe["age"].mode())
#measure of dispersion
print('var for df',dataframe["age"].var())
print('std for df',dataframe["age"].std())
print('rang for df',dataframe["age"].max() - dataframe["age"].min())
```

**Skewness** - describes the lack of symmetry in the mean (normal distribution = 0)

1. Skewness is a measure of symmetry, or more precisely, the lack of symmetry.
2. It is the degree of distortion from the symmetrical bell curve or the normal distribution.
3. **Positive Skewness** means when the tail on the right side of the distribution is longer or fatter. The mean and median will be greater than the mode.
4. **Negative Skewness** is when the tail of the left side of the distribution is longer or fatter than the tail on the right side. The mean and median will be less than the mode.
5. If the skewness is between -0.5 and 0.5, the data are fairly symmetrical
6. If the skewness is between -1 and -0.5(negatively skewed) or between 0.5 and 1(positively skewed), the data are moderately skewed
7. If the skewness is less than -1(negatively skewed) or greater than 1(positively skewed), the data are highly skewed

```
img1 = mpimg.imread('skew.png') 
plt.imshow(img1)
```

```
dataframe["age"].skew()
```

**Kurtosis** - 

1. Kurtosis is all about the tails of the distribution  not the peakedness or flatness.
2.Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution
3. High kurtosis in a data set is an indicator that data has heavy tails or outliers.
4. Low kurtosis in a data set is an indicator that data has light tails or lack of outliers. (Good one!)
5. **Mesokurtic:** This distribution has kurtosis statistic similar to that of the normal distribution. the standard normal distribution has a kurtosis of three
6. **Leptokurtic (Kurtosis > 3):** Distribution is longer, tails are fatter. Peak is higher and sharper than Mesokurtic, which means that data are heavy-tailed or profusion of outliers.
7. **Platykurtic: (Kurtosis < 3):** Distribution is shorter, tails are thinner than the normal distribution. The peak is lower and broader than Mesokurtic, which means that data are light-tailed or lack of outliers.

```
img2 = mpimg.imread('kurtosis.png') 
plt.imshow(img2)
```

```
dataframe["age"].kurt()
```

**Measure of Spread**

variability in the data  and How well the data is distributed? To get the overall description of the data, we will use central tendency and measure of description


1. **Range** - How well the data is spreadout? (Highest and lowest)
2. **Quartiles** are special percentiles, which divide the data into quarters. The first quartile, Q1, is the same as the 25th percentile, and the third quartile, Q3, is the same as the 75th percentile. The median is called both the second quartile, Q2, and the 50th percentile.
The next lowest 25% of numbers (up to the median).
The second highest 25% of numbers (above the median).
The highest 25% of numbers.
3. The **IQR** is a number that indicates how spread the middle half (i.e. the middle 50%) of the dataset is and can help determine outliers. It is the difference between the Q3 and Q1.
3. **Variance** -The average of the squared differences from the Mean (rough idea of spread from mean)
4. **Standard Deviation** - Square root of variance (more accurate spread)

```
#create a random list
a1 = np.random.randn(9)
print(a1)
```

```
#range from the above random list
np.max(a1) - np.min(a1)
```

```
img3 = mpimg.imread('iqr.png') 
plt.imshow(img3) 
```

```
#quartile
#1st quartile - 25%
#Q1 is the "middle" value in the first half of the rank-ordered data set.
q1 = np.percentile(a1, 25)
print(q1)
```

```
#2nd quartile - 50%
#median
q2 = np.percentile(a1, 50)
print(q2)
```

```
#3rd quartile - 75%
#Q3 is the "middle" value in the second half of the rank-ordered data set.
q3 = np.percentile(a1, 75)
print(q3)
```

```
#Interquartile range (IQR) = q3 - q1
IQR = q3 - q1
print(IQR)
```

```
a1_sample = np.random.choice(a1, 4)
print(a1_sample)
```

```
#Variance
print('Variance of population', np.var(a1))
print('Varaince of sample', np.var(a1_sample))  
```

```
#Standard Deviation
print('Standard Deviation of population', np.std(a1))
print('Standard Deviation of sample', np.std(a1_sample))  
```
