**importing library**

```
import pandas as pd
import numpy as np
```

```
#creating a list
mylist = [5.4,6.1,1.7,99.8]
mylist
```

```
#converting list to array
myarray = np.array(mylist)
myarray
```

```
#dictionary
mydict = {'w':10,'x':20,'y':30,'z':40}
mydict
```

**creating series**

```
#Pandas Series is a one-dimensional labeled array capable of holding data of any type
#nothing but a column in an excel sheet.
myseries1 = pd.Series(data=mylist)
print(myseries1)
myseries2 = pd.Series(data=myarray)
print(myseries2)
```

```
myseries = pd.Series(data=mydict)
myseries
```

```
# we access individual entries the same way as with lists and arrays
print(myseries1[2])
print(myseries2[3])
```

```
# we can add labels to the entries of a series
mylabels = ['first','second','third','fourth']
mylabels
```

```
myseries3 = pd.Series(data=mylist,index=mylabels)
print(myseries3)
```

```
#other method for embedding label with list/array
# we need not be explicit about the entries of pd.Series
myseries4 = pd.Series(myarray,mylabels)
print(myseries4)
```

```
# we can also access entries using the index labels
print(myseries4['second'])
```

```
# we can do math on series 
#empty is considered as NaN
myseries5 = pd.Series([5.5,1.1,8.8,1.6],['first','third','fourth','fifth'])
print(myseries5)
print('')
print(myseries5+myseries4)
```

**Dataframes**

```
#A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.
# we can combine series to create a dataframe using the concat function
df1 = pd.concat([myseries4,myseries5],axis=1,sort=False)
df1
```

**Dataframes combine (merge, concat, join)**

```
df7 = pd.DataFrame({"customer":['101','102','103','104'], 
                    'category': ['cat2','cat2','cat1','cat3'],
                    'important': ['yes','no','yes','yes'],
                    'sales': [123,52,214,663]},index=[0,1,2,3])
df7
```

```
df8 = pd.DataFrame({"customer":['101','103','104','105'], 
                    'color': ['yellow','green','green','blue'],
                    'distance': [12,9,44,21],
                    'sales': [123,214,663,331]},index=[4,5,6,7])
df8
```

```
pd.merge(df7,df8)
```

```
pd.merge(df7,df8, how='outer')
```

```
pd.merge(df7,df8, how='outer', on='customer')
```

```
#without sorting operation - column
pd.concat([df7,df8],axis=0,sort=False)
```

```
#with sorting - column
pd.concat([df7,df8],axis=0,sort=True)
```

```
#without sorting - row
pd.concat([df7,df8],axis=1,sort=False)
```

**Join can be used to combine columns of 2 dataframes that have different index values into a signle dataframe
The one difference between merge and join is that, merge uses common columns to combine two dataframes, whereas join uses the row index to join two dataframes**

**merge() for combining data on common columns or indices**

**.join() for combining data on a key column or an index**

**concat() for combining DataFrames across rows or columns**

```
daf3 = pd.DataFrame({'Q1': ['101', '102', '103'],
                     'Q2': ['201', '202', '203']},
                      index=['I0', 'I1', 'I2']) 

daf3
```

**Missing datas**

```
dataframe = pd.DataFrame({'Cricket':[1,2,np.nan,4,6,7,2,np.nan],
                  'Baseball':[5,np.nan,np.nan,5,7,2,4,5],
                  'Tennis':[1,2,3,4,5,6,7,8]})
```

```
dataframe
```

```
#dropping the missing values
dataframe.dropna()
```

```
# Use axis=1 for dropping columns with nan values
dataframe.dropna(axis=1)       
```

```
#particular row has two NaN- drop it
dataframe.dropna(thresh=2)
```

```
#filling the values for NaN
dataframe.fillna(value=0)
```

```
#filling NaN values as the mean value
dataframe['Baseball'].fillna(value=dataframe['Baseball'].mean())
```

**Groupby**

```
# Create dataframe as given below
dataframe1 = {'CustID':['1001','1001','1002','1002','1003','1003'],
       'CustName':['UIPat','DatRob','Goog','Chrysler','Ford','GM'],
       'Profitinlakhs':[2005,3245,1245,8765,5463,3547]}
```

```
dataframe1
```

```
#converting to dataframe
dataframe2 = pd.DataFrame(dataframe1)
```

```
dataframe2
```

**We can now use the .groupby() method to group rows together based on a column name. For example let's group based on CustID. This will create a DataFrameGroupBy object:**

```
dataframe2.groupby('CustID')
```

```
#This object can be saved as a variable
CustID_grouped = dataframe2.groupby("CustID")
```

```
#Now we can aggregate using the variable
CustID_grouped.mean()
```

```
#another method
dataframe2.groupby('CustID').mean()
```

```
#Standard deviation
CustID_grouped.std()
```

```
#min
CustID_grouped.min()
```

```
#max
CustID_grouped.max()
```

```
CustID_grouped.count()
```

```
CustID_grouped.describe()
```

```
CustID_grouped.describe().transpose()
```

```
CustID_grouped.describe().transpose()['1001']
```
