**More Matrix fun**

```python
sample_matrix = np.array(([50,200,5,10],[10,35,50,15],[25,100,145,120],[105,25,65,80]))
sample_matrix
```

```python
#Indexing rows
sample_matrix[1]
```

```python
# Getting an individual element value from the matrix - Method 1
sample_matrix[1][2]
```

```python
# Getting an individual element value from the matrix - Method 2
sample_matrix[1,2]
```

```python
#Slicing matrix
#Shape (3,3) from top right corner
sample_matrix[:3,1:]
```

```python
#Shape bottom row - Extract values from last row only
sample_matrix[3]
```

```python
#Shape bottom row - Including column selection (Alternate to above)
sample_matrix[3,:]
```

```python
#1st and 3rdcolumns
sample_matrix[:,(1,3)]
```

```python
#Allows to retrieve in any order - same 1st and 3rd columns
sample_matrix[:,(3,1)]
```

```python
#boolean function
sample_matrix <10
```

```python
#instead of boolean if i need values
boolean_array = sample_matrix <10
boolean_array
```

```python
sample_matrix[boolean_array]
```
